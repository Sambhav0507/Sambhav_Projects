DATA AUGMENTATION

Data augmentation occurs when you create new data based on modifications.Data augmentation on images would 
include transformations like:
• Flipping the image either horizontally or vertically
• Rotating the image
• Zooming in or out on the image
• Cropping the image
• Varying the color on the image

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import random 
import os
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
#this function will plot images in the form of a grid with 1 row and 10 columns where images are placed
def plotImages(images_arr):
    fig, axes= plt.subplots(1, 10, figsize=(20,20))
    axes= axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
gen = ImageDataGenerator( rotation_range=30, width_shift_range=0.5,
                        height_shift_range=0.1, shear_range=0.30, zoom_range=0.5,
                        channel_shift_range=30., horizontal_flip=True )
```

```python
chosen_image = random.choice(os.listdir('dogs vs cats/train/dog'))
```

```python
image_path = 'dogs vs cats/train/dog/' + chosen_image
```

```python
assert os.path.isfile(image_path)
```

```python
#obtain image
image = np.expand_dims(plt.imread(image_path),0)
plt.imshow(image[0])
```

```python
#generate batch of augmented images from this image
aug_iter = gen.flow(image)
```

```python
#get 10 samples of augmented images
aug_images= [next(aug_iter)[0].astype(np.uint8) for i in range(20)]
```

```python
#Augmented Images
plotImages(aug_images)
```
